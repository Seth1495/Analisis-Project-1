Analisis Project
Pada project dataset PT Sejahtera Bersama akan dilakukan ekplorasi dan analisis data untuk mendapatkan insight yang menjadi rekomendasi bisnis 

Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load dataset

```python
df = pd.read_csv("taskuse.csv")
df.head()
```

```python
df.shape
```

Exploratory Data Analysis

```python
dfbar = df["category_name"].value_counts()
print(dfbar)
plt.barh(data=dfbar, y=dfbar.index, width=dfbar.values, color = "orange")
plt.gca().invert_yaxis()
plt.show()
```

```python
dfbar7 = df[df['category_name'] == 'eBooks'].groupby('product_name').size().sort_values(ascending=False).head(10)
print(dfbar7)
plt.barh(data=dfbar7, y=dfbar7.index, width=dfbar7.values, color = "orange")
plt.gca().invert_yaxis()
plt.show()
```

```python
dfbar8 = df[df['category_name'] == 'Robots'].groupby('product_name').size().sort_values(ascending=False)
plt.barh(data=dfbar8, y=dfbar8.index, width=dfbar8.values, color = "orange")
plt.gca().invert_yaxis()
plt.show()
```

```python
dfbar1 = df.groupby('custcity')["total_sales"].sum().sort_values(ascending=False).head(10)
print(dfbar1)
plt.figure(figsize=(12,4))
sns.barplot(x= dfbar1.index, y=dfbar1, palette='pastel')
plt.xlabel('Position')
plt.ylabel('Overall AVG')
plt.title('Overall by Position')
```

```python
dfbar6 = df.groupby('custcity')["total_sales"].sum().sort_values(ascending=False).tail(10)
print(dfbar6)
plt.figure(figsize=(12,4))
sns.barplot(x= dfbar6.index, y=dfbar6, palette='pastel')
plt.xlabel('Kota')
plt.ylabel('Total Penjualan')
plt.title('Kota Total Penjualan Terendah')
```

```python
dfbar3 = df["product_name"].value_counts().head(10)
print(dfbar3)
plt.barh(data=dfbar3, y=dfbar3.index, width=dfbar3.values)
plt.gca().invert_yaxis()
plt.show()
```

```python
dfbar4 = df[df['custcity'] == 'Washington'].groupby('category_name').size().sort_values(ascending=False)
plt.figure(figsize=(12,4))
sns.barplot(x= dfbar4.index, y=dfbar4, palette='pastel')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category in Washington')
print(dfbar4)
```

```python
dfpie5 = df[(df['custcity'] == 'Washington') & (df["category_name"] == "eBooks")].groupby('product_name').size().sort_values(ascending=False).head(5)
dfpie5
```

```python
labels = ["Polar Robots", "Spherical Robots", "Articulated Robots", "GPS Drones", "Photograph Drones"]
plt.pie(dfpie5, labels=labels, autopct='%.1f%%')
plt.show()
```

Dari Category eBooks yang terjual di Kota Washington Produk spherical, polar, photographs drone, Gps drone, Articulated Robots menjadi produk yang mendominasi. Oleh karena itu rekomendasi bisnis yang dapat diberikan adalah :
